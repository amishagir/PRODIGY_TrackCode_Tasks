# Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
#Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read data set

df = pd.read_csv("HRDataset_v14.csv")
```

```python
df.drop(["DateofTermination"], axis = 1, inplace = True) 
```

```python
df1 = df.fillna(method='ffill')
df1
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
gender_employment_source_counts = df1.groupby(['Sex', 'EmploymentStatus', 'RecruitmentSource']).size().unstack()


fig, fig = plt.subplots(figsize=(10, 6))
gender_employment_source_counts.plot(kind='bar', stacked=True, ax=fig)

fig.set_xlabel('Gender and Employment Status')
fig.set_ylabel('Count')
fig.set_title('Distribution of Employment Status, Recruitment Source by Gender')
fig.legend(bbox_to_anchor=(1.05, 1))

plt.tight_layout()
     
```
